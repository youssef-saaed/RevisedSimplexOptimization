```python
from scipy.optimize import linprog
from lpp_optimize import RevisedSimplexOptimize
```

```python
c = [-2, -3]
A_ub , b_ub = [[2, 1], [1, 2]] , [4, 5]
A_eq , b_eq = None , None
bounds = [(0, None), (0, None)]
res = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method='revised simplex')
print(f"Optimal solution: {res.x}\n Optimal value: {-res.fun}")
```

```python
problem_1 = RevisedSimplexOptimize(c, A_ub, b_ub, A_eq, b_eq)
problem_1.solve()
```
